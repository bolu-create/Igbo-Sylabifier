```python
import re
#import pandas as pd

def syllabic_structure(sents):
    
    
    
        sent= sents.split() # Add error handling here also # Add cleaning of the sentence or something to ensure that the sentence that gets here is Igbo

        consonants= ['b','c','d','f','g','h','j','k','l','m','n','p','r','s','t','v','w','y','z','C','N']
        consonants2= ['b','c','d','f','g','h','j','k','l','m','n','p','r','s','t','v','w','y','z']


        vocab_struc= []
        x= ''

        for word in sent:

            #print (word)

            x = ''

            i = 0
            while i < len(word):
                if i != len(word) - 1:
                    if word[i:i+2] == 'ch':
                        x += 'C'
                        i += 2
                        continue

                    elif word[i:i+2] == 'gb':
                        x += 'C'
                        i += 2
                        continue

                    elif word[i:i+2] == 'gh':
                        x += 'C'
                        i += 2
                        continue

                    elif word[i:i+2] == 'gw':
                        x += 'C'
                        i += 2
                        continue

                    elif word[i:i+2] == 'kp':
                        x += 'C'
                        i += 2
                        continue

                    elif word[i:i+2] == 'kw':
                        x += 'C'
                        i += 2
                        continue

                    elif word[i:i+2] == 'nw':
                        x += 'C'
                        i += 2
                        continue

                    elif word[i:i+2] == 'ny':
                        x += 'C'
                        i += 2
                        continue

                    elif word[i:i+2] == 'sh':
                        x += 'C'
                        i += 2
                        continue

                x += word[i]
                i += 1

                #print(word)

            if len(word) > 2:
                if x != '':

                    for i in range(len(x)):

                        if i == len(x)-1 and x[i]== 'm':
                            x= x.replace(x[i],'N',1)

                        if (i == 0 or i != len(x)-1) and (x[i]== 'm' or x[i]=='n') and x[i+1] in consonants:
                            x= x.replace(x[i],'N',1)                

                        elif (i == 0 or i != len(x)-1) and (x[i]== 'm' or x[i]=='n'):
                            x= x.replace(x[i],'C',1)
                else:

                    for i in range(len(word)):
                        
                        if i == len(word)-1 and word[i]== 'm':
                            word= word.replace(word[i],'N',1)

                        elif (i == 0 or i != len(word)-1) and (word[i]== 'm' or word[i]=='n') and word[i+1] in consonants:
                            word= word.replace(word[i],'N',1)

                        elif (i == 0 or i != len(word)-1) and (word[i]== 'm' or word[i]=='n'):
                            word= word.replace(word[i],'C',1)

            if x != '':

                for i in range(len(x)): 

                    if x[i] not in consonants:
                        #word[i]= "V"
                        x= x.replace(x[i],'V')

                for i in range(len(x)): 

                    if x[i] in consonants2:
                        #word[i]= "V"
                        x= x.replace(x[i],'C')

                vocab_struc.append(x)

            else:
                for i in range(len(word)): 

                    if word[i] not in consonants:
                        #word[i]= "V"
                        word= word.replace(word[i],'V')

                for i in range(len(word)): 

                    if word[i] in consonants2:
                        #word[i]= "V"
                        word= word.replace(word[i],'C')

                vocab_struc.append(word)
                
        # Corrected code for syllabicator
        def vocabs_from_pattern(y):
            el = list(y)
            length= len(el)

            for i in range(length):
                if el[i] == 'V':
                    el[i] = 'V-' if i != length - 1 else 'V'
                elif el[i] == 'N':
                    el[i] = 'N-' if i != length - 1 else 'N'
                elif el[i] == 'C':
                    if el[i + 1] == 'V':
                        el[i] = 'C'
                    else:
                        el[i] = 'C-'

            return ''.join(el)

       
        #return ' '.join(vocab_struc)#,list(zip(sent,vocab_struc))
        x= vocabs_from_pattern(' '.join(vocab_struc))#,list(zip(sent,vocab_struc))


        
        mappings = {'ch': 'C', 'gb': 'C', 'gh': 'C', 'gw': 'C', 'kp': 'C', 'kw': 'C', 'nw': 'C', 'ny': 'C', 'sh': 'C'}
        digraphs = ['ch', 'gb', 'gh', 'gw', 'kp', 'kw', 'nw', 'ny', 'sh']
        word= sents
        syllable= x
        indexes= [i for i, char in enumerate(syllable) if char == '-']
        word_list= list(word)
        word_syllable=''
        di=[]
        output_list = []
        skip_next = False

        def replace_x_with_y(x, y):
            result = []
            y_index = 0

            for char in x:
                if char == 'C' and y_index < len(y):
                    result.extend(y[y_index])
                    y_index += 1
                else:
                    result.append(char)

            return result


        if len(word) == len(re.sub(r"[^\w\s]","", syllable)):

            for index in indexes:
                word_list.insert(index,'-')
            word_syllable=''.join(word_list)

        else:

            for i in range(len(word)):
                if word[i:i+2] in digraphs:
                    #print(word[i:i+2])
                    di.append(word[i:i+2])
                    word_list[i] = mappings.get(word[i:i+2], word)

            for i, item in enumerate(word_list):
                if skip_next:
                    skip_next = False
                    continue
                output_list.append(item)
                if item == 'C':
                    skip_next = True

            for index in indexes:
                    output_list.insert(index,'-')
                    word_syllable=''.join(output_list)

            output_list= replace_x_with_y(output_list, di)
            word_syllable=''.join(output_list)


        return word_syllable#,x

```
