### Syllabic Distribution

```python
import matplotlib.pyplot as plt

def count_syllables(word):
    # Define Igbo syllable counting rules
    #vowels = ['a', 'e', 'i', 'o', 'u']
    consonants= ['b','c','d','f','g','h','j','k','l','m','n','p','r','s','t','v','w','y','z']
    syllables = 0

    # Count syllables based on vowel-consonant patterns
    for i in range(len(word)):
        #if word[i] in vowels:
        if word[i] not in consonants:
            #if i == len(word) - 1 or word[i] in vowels:
            if i == len(word) - 1 or word[i] not in consonants:
                syllables += 1

    return syllables

def syllable_distribution(corpus):
    syllable_counts = {}
    for word in corpus:
        syllables = count_syllables(word)
        if syllables in syllable_counts:
            syllable_counts[syllables] += 1
        else:
            syllable_counts[syllables] = 1
    return syllable_counts

# Example Corpus (replace 'corpus' with your actual Igbo corpus)
# corpus = ['akụ', 'alụ', 'ogwe', 'okwukwe', 'nwa', 'ife', 'ifee', 'ọfọ']
#corpus = ['buonana']

corpus= open("my_corpus.txt", encoding= "utf8")
corpus= corpus.read()
corpus= corpus.split()
# Compute Word Frequencies
word_freq = Counter(corpus)
# Rank the Words
ranked_words = sorted(word_freq.items(), key=lambda x: x[1], reverse=True)
corpus= [x for x, y in ranked_words]

# Count Syllables and Calculate Distribution
distribution = syllable_distribution(corpus)

# Plot the Distribution
syllable_counts = sorted(distribution.items())
syllables, counts = zip(*syllable_counts)


plt.bar(syllables, counts)
plt.xlabel('Number of Syllables')
plt.ylabel('Frequency')
plt.title('Syllable Distribution in the Igbo Corpus')
plt.show()


#syllables
#counts
#syllable_counts
```

- If m occurs at the end of a word it is a syllabic nassal
- If n or m precede a consonant, when within a word, it is a nasal
