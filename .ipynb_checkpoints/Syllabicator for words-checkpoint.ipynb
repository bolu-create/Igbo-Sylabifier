```python
# Optimization of my code
def vocabs_from_pattern(y):
    el = list(y)
    length= len(el)

    for i in range(length):
        if el[i] == 'V':
            el[i] = 'V-' if i != length - 1 else 'V'
        elif el[i] == 'N':
            el[i] = 'N-' if i != length - 1 else 'N'
        elif el[i] == 'C':
            if el[i + 1] == 'V':
                el[i] = 'C'
            else:
                el[i] = 'C-'

    return ''.join(el)

# Test the function
pattern = 'VCVCVCVNCVCCN'
result = vocabs_from_pattern(pattern)
print(result)  # Output: 'V-CV-CV-CVN-CVC-CN'

```
